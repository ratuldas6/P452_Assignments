# Question 2

```python
data_raw = []

# importing data from assign3fit.txt
file = open("assign3fit.txt", "r+")

for line in file:
    x = line.split()
    y = []
    for i in range(len(x)):
        y.append(x[i])
    data_raw.append(y)

del data_raw[0]

X = np.array([float(x) for (x,y) in data_raw])
Y = np.array([float(y) for (x,y) in data_raw])
```

```python
print("The condition number for the matrix while using monomials as basis is :", mylibrary.poly_fit(X, Y, mylibrary.poly_basis)[1])
print()
print("The condition number for the matrix while using chebyshev polynomials as basis is :", mylibrary.poly_fit(X, Y, mylibrary.cheby_basis)[1])
```

```python
pol_par = mylibrary.poly_fit(X, Y, mylibrary.poly_basis)[0]
cheb_par = mylibrary.poly_fit(X, Y, mylibrary.cheby_basis)[0]

xlist1 = np.linspace (-0.5, 1.5, 100)
def f1(x):
    y = 0
    for i in range (0, pol_par.shape[0]):
        y += pol_par[i]*x**i
    return y

def f2(x):
    y = cheb_par[0] + cheb_par[1]*(2*x-1) + cheb_par[2]*(8*(x**2)-8*x+1) + cheb_par[3]*(32*(x**3)- 48*(x**2) + 18*x - 1) 
    return y
```

```python
plt.scatter(X,Y)
plt.plot(xlist1, f1(xlist1), '--', color='red')
plt.title("Using monomial as basis function")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.scatter(X,Y)
plt.plot(xlist1, f2(xlist1), '--', color='red')
plt.title("Using Chebyshev polynomials as basis function")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
